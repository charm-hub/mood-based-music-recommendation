```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
df = pd.read_csv('tracks.csv')
df.head()
```

```python
df = df[['names', 'artist_names', 'energy', 'danceability',
         'valences', 'tempos', 'acousticness', 'loudness']]

df.dropna(inplace=True)
df.head()
```

```python
def assign_mood(row):
  if row['energy'] > 0.75 and row['danceability'] > 0.75 and row['valences'] > 0.75:
    return 'happy'
  elif row['energy'] < 0.4 and row['valences'] < 0.4:
    return 'sad'
  elif row['energy'] > 0.85:
    return 'energetic'
  elif row['energy'] < 0.4 and row['acousticness'] > 0.6:
    return 'Calm'
  else:
    return 'Romantic'
```

```python
df['mood'] = df.apply(assign_mood, axis=1)

df[['names', 'artist_names', 'mood']].head(10)
```

```python
df['mood'].value_counts()
```

```python
def recommend_by_mood(mood, n=10):
    mood_songs = df[df['mood'] == mood]
    return mood_songs.sample(n)[['names', 'artist_names']]
```

```python
mood_dropdown = widgets.Dropdown(
    options=['happy', 'sad', 'energetic', 'Calm', 'Romantic'],
    description='Select Mood:'
)

display(mood_dropdown)
```

```python
def on_mood_change(change):
    print(f"\nðŸŽ§ Songs for {change['new']} Mood:\n")
    display(recommend_by_mood(change['new']))

mood_dropdown.observe(on_mood_change, names='value')
```

###  User Interface (UI)


```python
mood_dropdown = widgets.Dropdown(
    options=['happy', 'sad', 'energetic', 'Calm', 'Romantic'],
    value='happy',
    description='Mood:',
    style={'description_width': 'initial'}
)
```

```python
recommend_button = widgets.Button(
    description='Get Recommendations ðŸŽµ',
    button_style='success'
)
```

```python
output = widgets.Output()
```

```python
def on_button_click(b):
    with output:
        clear_output()
        mood = mood_dropdown.value
        print(f"ðŸŽ§ Recommended Songs for {mood} Mood:\n")
        display(recommend_by_mood(mood))

recommend_button.on_click(on_button_click)
```

```python
display(
    widgets.VBox([
        widgets.HTML("<h2>ðŸŽ¶ Mood-Based Music Recommendation System</h2>"),
        mood_dropdown,
        recommend_button,
        output
    ])
)
```
